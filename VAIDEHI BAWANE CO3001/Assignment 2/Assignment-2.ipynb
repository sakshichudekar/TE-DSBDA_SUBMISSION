# Fill Null Values with mMean and Mode

```python
df2=df.copy()
```

```python
for col in df.columns:
    if col in df2.columns: 
        if df2[col].isnull().any() and pd.api.types.is_numeric_dtype(df2[col]):
            df2[col]=df2[col].fillna(df2[col].mean())
        
```

```python
print(df2.isnull().sum())
```

```python
df2.info
```

```python
if 'City' in df2.columns:
    if df2['City'].isnull().any():
        mode_value = df2['City'].mode()
        if not mode_value.empty:
            most_frequent_city = mode_value[0]
            df2['City'] = df2['City'].fillna(most_frequent_city)
        else:
            print("No mode found!")
```

```python
df2.info
```

# Fill null values with forward fill values

```python
df2=df.copy()
```

```python
for col in df2.columns:
    if df2[col].isnull().any():
        df2[col]=df2[col].ffill()
df2.isnull().sum()
df2
```

# Fill null values using backward fill values

```python
df3=df.copy()
for col in df3.columns:
    if df3[col].isnull().any():
        df3[col]=df3[col].bfill()
df3.isnull().sum()
df3
```
